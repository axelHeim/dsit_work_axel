```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import pandas as pd
import math
```

```python
workdir = Path("/nfs/dust/belle2/user/axelheim/BranchSeparator/BsX_hadrIDtrials/data/TSD_BsX_15k_each_6trees_only4mom_noEnergy")
#workdir = Path("/nfs/dust/belle2/user/axelheim/BranchSeparator/run_BsXHc/data/TSD_BsXHc_15k_each_6trees_only4mom")
workdir.mkdir(parents=True, exist_ok=True)
leaves = np.load(workdir / 'leaves.npy')
SA = np.load(workdir / 'is_left_arr.npy')
global_tag = np.load(workdir / 'global_tag.npy')

#pion momenta all modes

pi = []
K = []
gamma = []
momentum = []
px = []
py = []
pz = []
mode_list = []
bg = []
label = []

for i in range(len(leaves)):
    for j in range(len(global_tag[1])):
        tag = global_tag[i][j]
        # filter for non deleted FSPs and for pion OR Kaon OR gamma
        if str(tag).find("DELETED") == -1: 
            if (str(tag).find("pi") != -1 or str(tag).find("K") != -1) or str(tag).find("gamma") != -1:
                px.append(leaves[i][j][0])
                py.append(leaves[i][j][1])
                pz.append(leaves[i][j][2])
                momentum.append(np.sqrt((leaves[i][j][0:3]**2).sum(axis=0))) # this is the momentum
                mode_list.append(int(str(global_tag[i,-1])[2]))
                label.append(SA[i,j])
                
                pi.append(0)
                K.append(0)
                gamma.append(0)
                bg.append(0)
                if str(tag).find("BG") != -1:
                    bg[-1] = 1

                if str(tag).find("pi") != -1 :
                    pi[-1] = 1

                if str(tag).find("K") != -1 :
                    K[-1] = 1

                if str(tag).find("gamma") != -1 :
                    gamma[-1] = 1




```

```python
df = pd.DataFrame({"bg" : bg,
                   "mode" : mode_list,
                   "momentum" : momentum,
                   "K" : K,
                   "pi" : pi,
                   "gamma" : gamma,
                   "px":px,
                   "py":py,
                   "pz":pz,
                   "label" : label})

K_df = df[df.K == 1]
noBG_df = df[df.bg == 0]
```

```python
bins = np.linspace(0, 3750, 100)
fig, axs = plt.subplots(3,2)
fig.suptitle('Label specific momentum spectra BsXHc particles')
fig.set_size_inches(16, 10)

for mode in range(6):
    #print("mode:",mode)
    axs_index0 = math.floor(mode/2)
    axs_index1 = mode % 2
    axs[axs_index0,axs_index1].hist(df[(df.label==0) & (df['mode']==mode)].momentum, bins, alpha=0.5, label='X')
    axs[axs_index0,axs_index1].hist(df[(df.label==1) & (df['mode']==mode)].momentum, bins, alpha=0.5, label='Bs')
    axs[axs_index0,axs_index1].hist(df[(df.label==2) & (df['mode']==mode)].momentum, bins, alpha=0.5, label='background')
    #axs[axs_index0,axs_index1].hist(df[(df.label==3) & (df['mode']==mode)].momentum, bins, alpha=0.5, label='Hc')
    axs[axs_index0,axs_index1].set_title("mode: {}".format(mode))
    axs[axs_index0,axs_index1].legend(loc='upper right')

fig.savefig('label_momSpectra_BsXHc.pdf')
#plt.show()
```
