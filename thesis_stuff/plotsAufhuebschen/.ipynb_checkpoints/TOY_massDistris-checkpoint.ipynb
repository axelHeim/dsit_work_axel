```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
```

```python
import tensorflow as tf
import tensorflow_probability as tfp
```

```python
n_events = 100000
min_mass = 0
max_mass= 2000
width_cast = tf.cast(148.0, tf.float64)
mass_cast = tf.cast(770.0, dtype=tf.float64)

particle_mass_rho = tf.broadcast_to(mass_cast, shape=(n_events,))
#if width_cast > 0: # comparison with tf. variable is not working
particle_mass_rho = tfp.distributions.TruncatedCauchy(loc=mass_cast,
                                                    scale=width_cast,
                                                    low=min_mass,
                                                    high=max_mass).sample(n_events)
```

```python
n_events = 100000
min_mass = 1000
max_mass= 3000
width_cast = tf.cast(2.0, tf.float64)
mass_cast = tf.cast(2007.0, dtype=tf.float64)

particle_mass_D0st = tf.broadcast_to(mass_cast, shape=(n_events,))
#if width_cast > 0: # comparison with tf. variable is not working
particle_mass_D0st = tfp.distributions.TruncatedCauchy(loc=mass_cast,
                                                    scale=width_cast,
                                                    low=min_mass,
                                                    high=max_mass).sample(n_events)
```

```python
print(type(particle_mass_rho))
proto_tensor = tf.make_tensor_proto(particle_mass_rho)  # convert `tensor a` to a proto tensor
particle_mass_rho_np = tf.make_ndarray(proto_tensor)
print(type(particle_mass_rho_np))
```

```python
print(type(particle_mass_D0st))
proto_tensor = tf.make_tensor_proto(particle_mass_D0st)  # convert `tensor a` to a proto tensor
particle_mass_D0st_np = tf.make_ndarray(proto_tensor)
print(type(particle_mass_D0st_np))
```

```python
fig, ax = plt.subplots()

bins = np.linspace(0, 2000, 100)

#fig.suptitle(r'{}'.format("summed_wrongPred_p"))
fig.set_size_inches(8, 5)

ax.set_xlabel(r'mass / MeV')
ax.set_ylabel("density")


ax.hist(particle_mass_rho_np,
            bins=bins,
            density=True, alpha=0.8, label=(r'$\rho$ meson'),histtype='step')

#ax.legend()

plt.grid()
plt.savefig('plots/Rho_Distri.pdf')
plt.show()
```

```python
fig, ax = plt.subplots()

bins = np.linspace(1960, 2060, 100)

#fig.suptitle(r'{}'.format("summed_wrongPred_p"))
fig.set_size_inches(8, 5)

ax.set_xlabel(r'mass / MeV')
ax.set_ylabel("density")


ax.hist(particle_mass_D0st_np,
            bins=bins,
            density=True, alpha=0.8, label=(r'$D^{*0}$ meson'),histtype='step')

#ax.legend()

plt.grid()
plt.savefig('plots/D0st_Distri.pdf')
plt.show()
```
