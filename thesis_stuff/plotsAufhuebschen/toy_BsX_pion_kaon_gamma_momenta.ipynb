```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import pandas as pd
import math
```

```python
nfs_path = "/nfs/dust/belle2/user/axelheim/thesisPlots_data/"
df = pd.read_csv(nfs_path +  "toy_BsX_pion_kaon_gamma_momenta.csv")
```

```python
K_df = df[df.K == 1]
noBG_df = df[df.bg == 0]
```

```python
bins = np.linspace(0, 3750, 100)
fig, axs = plt.subplots(3,2)
fig.suptitle('')
fig.set_size_inches(16, 10)


fontsize = 10

for mode in range(6):
    #print("mode:",mode)
    axs_index0 = math.floor(mode/2)
    axs_index1 = mode % 2
    axs[axs_index0,axs_index1].hist(noBG_df[(noBG_df.pi==1) & (noBG_df['mode']==mode)].momentum, bins, alpha=0.5, label='pions')
    axs[axs_index0,axs_index1].hist(noBG_df[(noBG_df.K==1) & (noBG_df['mode']==mode)].momentum, bins, alpha=0.5, label='kaons')
    axs[axs_index0,axs_index1].hist(noBG_df[(noBG_df.gamma==1) & (noBG_df['mode']==mode)].momentum, bins, alpha=0.5, label='photons')
    axs[axs_index0,axs_index1].set_title("mode {}".format(mode))
    axs[axs_index0,axs_index1].legend(loc='upper right')
    
    label = axs[axs_index0,axs_index1].set_xlabel(r'$p$ / MeV', fontsize = fontsize)
    axs[axs_index0,axs_index1].set_ylabel(r'particles', fontsize = fontsize)
    axs[axs_index0,axs_index1].xaxis.set_label_coords(1.05, -0.025)

fig.savefig('plots/pi_K_gamma_momSpectra_noBG_BsX.pdf')
#plt.show()

#r'$\sin (x)$'
```
