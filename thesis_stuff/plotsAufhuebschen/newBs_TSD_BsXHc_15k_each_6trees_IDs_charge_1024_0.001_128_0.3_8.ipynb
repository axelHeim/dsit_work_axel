```python
def curve_plot_accuracies(plot_path, plot_name, y_label, title,
                         acc_all, acc_bg, acc_X=None, acc_Bt=None, acc_Bs=None,  acc_Hc=None): 
    #title=(data_name + '/' + spec)
    #ylabel="Accuracy/Validation"
    font = 20
    #plot_name = "accuracies_" + plotname_extension

    name=plt.figure(figsize=(16,10))#figsize=(3.487,3.487/1.618))
    ax =plt.gca()

    #set ylim lower bound to zero
    plt.gca().set_ylim(bottom=0)
    #plt.gca().set_xlim(left=0)
    
    print(type(acc_all))
    print(type(acc_bg))
    print(type(acc_X))
    print(type(acc_Bt))
    print(type(acc_Bs))
    

    acc_all.plot(x="Step", y="Value", color='b', ax=ax, alpha=1, label="accuracy all")
    
    

    
    acc_bg.plot(x="Step", y="Value", color='g', ax=ax, alpha=1, label="background" )
    
    if isinstance(acc_X, pd.core.series.Series):
        acc_X.plot(x="Step", y="Value", color='r', ax=ax, alpha=1, label="X" )
        
    if isinstance(acc_Bt, pd.core.series.Series):
        acc_Bt.plot(x="Step", y="Value", color='r', ax=ax, alpha=1, label="B-tag" )
    
    if isinstance(acc_Bs, pd.core.series.Series):
        acc_Bs.plot(x="Step", y="Value", color='y', ax=ax, alpha=1, label="B-sig" )
    
    if isinstance(acc_Hc, pd.core.series.Series):
        acc_Hc.plot(x="Step", y="Value", color='c', ax=ax, alpha=1, label="Hc" )

    name= plt.xlabel('Epochs', fontsize=font)
    name= plt.ylabel(y_label, fontsize=font)
    name= plt.title(title, fontsize=font)
    name = plt.xticks(fontsize=font, rotation=360)
    name =  plt.yticks(fontsize=font, rotation=360)
    name= plt.legend(fontsize=font)

    plt.savefig(plot_path/ (plot_name + '.pdf'))
```
