```python
# run with localhost but only conda base

from decaylanguage import DecFileParser
from decaylanguage import DecayChainViewer

```

```python
dec_path = "/afs/desy.de/user/a/axelheim/private/MC_studies/my6modes/decfiles/"
for modeNum in range(6):
    parser = DecFileParser(dec_path + f'mode{modeNum}.dec')
    parser.parse()

    # Inspect what decays are defined
    parser.list_decay_mother_names()
    d = parser.build_decay_chains('Upsilon(4S)')
    dcv = DecayChainViewer(d)
    dcv.graph
    dcv.graph.render(filename=f'plots/newBsig_mode{modeNum}', format='pdf', view=True, cleanup=True)
```
